# Gapminder 
GDP per capita vs life expectancy.

Source: https://bookdown.org/jamie/python_visualisation/building-graphs-with-full-control-1.html


```python
import pandas as pd

#import plotly.graph_objects as go
import plotly.express as px
```

```python
DATA_URL = "https://raw.githubusercontent.com/wcj365/python-dataviz/master/plotly/GapminderFiveYearData.csv"

gapminder = pd.read_csv(DATA_URL)
gapminder.head()
```

```python
fig = px.scatter(
    gapminder.sort_values('year'), 
    x='gdpPercap', y='lifeExp',
    color='continent', size='pop',
    #log_x=True,
    animation_frame='year',
    animation_group='country',
    hover_name='country',
    range_y=[0,100],
    range_x=[0, 50000],
    labels={
        'pop': 'Population',
        'gdpPercap': 'GDP per Capita',
        'year': 'Year',
        'lifeExp': 'Life Expectancy',
        'continent': 'Continent'
    },
    size_max=75
)

fig.update_layout(
    template='plotly_dark'
#  margin=dict(r=10, t=25, b=40, l=60)
)

fig.show()
```
